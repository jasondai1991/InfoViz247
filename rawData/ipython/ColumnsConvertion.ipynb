```python
import pandas as pd
df = pd.read_csv('./flightsOfYear.csv', sep=',', header=0)
```

```python
df.drop(['CANCELLATION_REASON_A','CANCELLATION_REASON_B','CANCELLATION_REASON_C','ARRIVAL_DELAY_BIN'],axis=1, inplace = True)
```

```python
def get_date_of_week(row):
    date = 1
    if row['DAY_OF_WEEK_2']==1:
        date = 2
    elif row['DAY_OF_WEEK_3']==1:
        date = 3
    elif row['DAY_OF_WEEK_4']==1:
        date = 4
    elif row['DAY_OF_WEEK_5']==1:
        date = 5
    elif row['DAY_OF_WEEK_6']==1:
        date = 6
    elif row['DAY_OF_WEEK_7']==1:
        date = 7
    return date

df['date_of_week'] = df.apply(get_date_of_week,axis=1)
```

```python
df.drop(['DAY_OF_WEEK_2', 'DAY_OF_WEEK_3', 'DAY_OF_WEEK_4', 'DAY_OF_WEEK_5',
       'DAY_OF_WEEK_6', 'DAY_OF_WEEK_7'],axis=1, inplace = True)
```

```python
def get_airline(row):
    airlines = ['AIRLINE_AS', 'AIRLINE_B6','AIRLINE_DL', 'AIRLINE_EV', 
                'AIRLINE_F9', 'AIRLINE_HA', 'AIRLINE_MQ','AIRLINE_NK', 
                'AIRLINE_OO', 'AIRLINE_UA', 'AIRLINE_US', 'AIRLINE_VX','AIRLINE_WN']
    answer = ""
    for airline in airlines:
        if row[airline]==1:
            answer = airline[8:]
            break
    if answer == "":
        answer = "AA"
    if answer == "VX":
        answer = "AS"
    if answer == "US":
        answer = "AA"
    return answer

df['airline'] = df.apply(get_airline,axis=1)
```

```python
df.drop(['AIRLINE_AS', 'AIRLINE_B6','AIRLINE_DL', 'AIRLINE_EV', 
                'AIRLINE_F9', 'AIRLINE_HA', 'AIRLINE_MQ','AIRLINE_NK', 
                'AIRLINE_OO', 'AIRLINE_UA', 'AIRLINE_US', 'AIRLINE_VX','AIRLINE_WN'], axis=1, inplace = True)
```

```python
df.SCHEDULED_DEPARTURE = df.SCHEDULED_DEPARTURE.astype("int")
df.SCHEDULED_ARRIVAL = df.SCHEDULED_ARRIVAL.astype("int")
```

```python
def straighten_time(row):
    row['SCHEDULED_DEPARTURE'] = get_time(row['SCHEDULED_DEPARTURE'])
    row['SCHEDULED_ARRIVAL'] = get_time(row['SCHEDULED_ARRIVAL'])
    return row

def get_time(time):
    hour = str(int(time /100))
    minute = str(time %100)
    if len(hour) < 2:
        hour = "0"+hour
    if len(minute) < 2:
        minute = "0"+ minute
    answer = hour + ":" + minute
    return answer

df2= df.apply(straighten_time,axis=1)
```

```python
df2.to_csv("allYearFlights.csv")
```

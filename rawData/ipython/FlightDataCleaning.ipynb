```python
df_raw = df[['ORIGIN_AIRPORT', 'DESTINATION_AIRPORT']]
```

```python
df_airport = pd.read_csv('./L_AIRPORT.csv', sep=',', header=0)
```

```python
df_airport[["City","Desc", "1", "2"]] = df_airport["Description"].str.split(",",expand = True) 
```

```python
df_airport[["State", "AirportName"]] = df_airport["Desc"].str.split(":",expand = True) 
```

```python
# final version of airport data without longitute, latitude
df_airport = df_airport.drop(['Description', 'Desc','1','2'], axis=1)
```

```python

import pandas as pd   
df2 = pd.read_csv('./airports.dat', sep ='\t', engine = 'python')
df2.drop(['airport-name'], axis=1)
```

```python
new_df = df2.join(df_airport.set_index('Code'), on='airport-id', how='inner')
```

```python
new_df = new_df[new_df['State'].str.strip().str.len()==2]
```

```python
final_airport = new_df.drop(['airport-name'],axis=1)
```

```python
final_airport.to_csv("USAirport.csv")
```

```python
departure_airport = final_airport.rename(index=str, columns={"latitude": "de-latitude", "longitude": "de-longitude"})
```

```python
departure_airport.drop(['City','State','AirportName'],axis=1,inplace=True)
```

```python
arrival_airport = departure_airport.rename(index=str, columns={"de-latitude": "arr-latitue", "de-longitude": "arr-longitude"})
```

```python
df_ts1 = df_raw.join(departure_airport.set_index('airport-id'), on='ORIGIN_AIRPORT', how='inner')
```

```python
df_ts2 = df_ts1.join(arrival_airport.set_index('airport-id'), on='DESTINATION_AIRPORT', how='inner')
```

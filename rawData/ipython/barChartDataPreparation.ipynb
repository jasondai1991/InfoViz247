```python
import pandas as pd

df2 = pd.read_csv('./allYearFlights.csv', sep=',', header=0, index_col=0)
```

```python
import math

def calc_bin(row):
    x = row['SCHEDULED_DEPARTURE']
    total = int(x[0:2])*100 + int(x[3:])
    return math.trunc(total/10)

df2['departure_bin'] = df2.apply(calc_bin,axis=1)
```

```python
df_airport = pd.read_csv('./L_AIRPORT.csv', sep=',', header=0)
df_airport[["City","Desc", "1", "2"]] = df_airport["Description"].str.split(",",expand = True) 
df_airport[["State", "AirportName"]] = df_airport["Desc"].str.split(":",expand = True) 
df_airport = df_airport.drop(['Description', 'Desc','1','2'], axis=1)
```

```python
df2.drop(['DAY_OF_YEAR','DESTINATION_AIRPORT',
       'SCHEDULED_DEPARTURE', 'DEPARTURE_DELAY', 'SCHEDULED_TIME', 'DISTANCE',
       'SCHEDULED_ARRIVAL', 'ARRIVAL_DELAY', 'DIVERTED', 'CANCELLED',
       'date_of_week'],axis=1, inplace = True)
```

```python
new_df = df2.join(df_airport.set_index('Code'), on='ORIGIN_AIRPORT', how='inner')
```

```python
new_df.drop(['State', 'AirportName','ORIGIN_AIRPORT'],axis=1,inplace=True)
```

```python
import numpy
by_city=new_df.groupby(['City','departure_bin']).size().reset_index(name='counts')
by_city['counts'] = numpy.trunc(by_city['counts']/40*100)/100
```

```python
by_city.to_csv("city_data.csv")
```

```python
by_city = by_city.pivot(index='City', columns='departure_bin', values='counts')
```

```python
by_city.fillna(0,inplace=True)
```

```python
by_city.astype("int",inplace=True)
```

```python
by_city.insert(0, 'Type', 'City')
```

```python
by_city.to_csv("city_data.csv")
```

```python
by_airline = new_df.groupby(['airline','departure_bin']).size().reset_index(name='counts')
```

```python
airline_map = {"NK":"Spirit","WN":"Southwest","B6":"Jetblue","UA":"United","AA":"American","DL":"Delta","OO":"SkyWest",
             "F9":"Frontier","EV":"ExpressJet","HA":"Hawaiian","AS":"Alaska","MQ":"AmericanEagle"};
```

```python
by_airline.replace({"airline": airline_map}, inplace = True)
```

```python
by_airline.insert(0, 'Type', 'Airline')
```
